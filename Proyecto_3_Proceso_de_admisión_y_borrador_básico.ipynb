```python
import pandas as pd

ruta_archivo = '/content/drive/MyDrive/Colab Data/applicant_list_7.txt'

nombres_columnas = ['nombre', 'apellido', 'EPH', 'ECH', 'EMA', 'ECS', 'ESP', 'OP1', 'OP2', 'OP3']

df = pd.read_csv(ruta_archivo, sep=' ', names=nombres_columnas)

df

```

```python
df['nombre_completo'] = df['nombre'] + ' ' + df['apellido']
df
```

```python
df['PROMEDIO'] = (df['EPH'] + df['ECH'] + df['EMA'] + df['ECS'] + df['ESP']) / 5
df
```

```python
df['SUMA'] = df['OP1'] + df['OP2'] + df['OP3']
df
```

```python
# definición de una función para calcular puntaje de admisión

def calcular_puntaje_admision(linea, opción):
  puntaje = 0
  departamento = linea[opción]
  if departamento == 'Physics':
    puntaje = (linea['EPH'] + linea ['EMA'] + linea ['ESP']) /3
  elif departamento == 'Mathematics':
    puntaje = (linea['ECH'] + linea ['ECS'] + linea ['ESP']) /3
  elif departamento == 'Engineering':
    puntaje = (linea['EMA'] + linea ['ECS'] + linea ['EPH'] + linea ['ESP']) /4
  elif departamento == 'Chemistry': # Corrected typo
    puntaje = (linea['ECH'] + linea ['EPH'] + linea ['ESP']) /3
  elif departamento == 'Biotech':
    puntaje = (linea['EMA'] + linea ['ECS'] + linea ['ECH'] + linea ['ESP'])  /4
  else:
    return None
  return round(puntaje, 2)

df['OP1_score'] = df.apply(lambda x:calcular_puntaje_admision(x, 'OP1'), axis=1)
df['OP2_score'] = df.apply(lambda x:calcular_puntaje_admision(x, 'OP2'), axis=1)
df['OP3_score'] = df.apply(lambda x:calcular_puntaje_admision(x, 'OP3'), axis=1)
display(df)
```

```python
N = int(input("Ingrese el cupo máximo por departamento: "))
departamento_seleccionado = input("Ingrese el departamento: ")
```

```python
aspirantes_departamento = df[(df['OP1'] == departamento_seleccionado)]
aspirantes_ordenados = aspirantes_departamento.sort_values(by='OP1_score', ascending=False)
tabla_aspirantes_adminitidos = aspirantes_ordenados.head(N)

# Ensure 'nombre_completo' is in the subset DataFrame
if 'nombre_completo' not in tabla_aspirantes_adminitidos.columns:
    tabla_aspirantes_adminitidos['nombre_completo'] = tabla_aspirantes_adminitidos['nombre'] + ' ' + tabla_aspirantes_adminitidos['apellido']


tabla_en_formato = pd.DataFrame(
    { 'nombre_completo': tabla_aspirantes_adminitidos['nombre_completo'],
      'departamento': tabla_aspirantes_adminitidos['OP1'],
      'puntaje': tabla_aspirantes_adminitidos['OP1_score'],
     }
)
display(tabla_en_formato)
aspirantes_departamento
```
